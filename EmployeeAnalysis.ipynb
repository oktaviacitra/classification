```python
import pandas as pd

dataset_path = "/content/drive/MyDrive/Learning Journey/Employee Analysis/dataset.csv"
dataframe = pd.read_csv(dataset_path)
dataframe.head(3)
```

```python
dataframe.isna().sum()
```

```python
dataframe.dtypes
```

```python
categorical_columns = dataframe.select_dtypes(include='object').columns.to_list()
numerical_columns = dataframe.select_dtypes(include='int64').columns.to_list()
```

```python
from sklearn.preprocessing import StandardScaler

dataframe[numerical_columns] = StandardScaler().fit_transform(dataframe[numerical_columns])
dataframe.head(3)
```

```python
from sklearn.preprocessing import OrdinalEncoder

dataframe[categorical_columns] = OrdinalEncoder().fit_transform(dataframe[categorical_columns])
```

```python
dataframe.head(3)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(dataframe)
```
